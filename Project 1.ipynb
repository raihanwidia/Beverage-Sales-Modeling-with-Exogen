```python
fig, ax = plt.subplots(figsize=(25, 5))

# store
df_transactions_store1 = store[store['family'] == 'AUTOMOTIVE']
df_transactions_store1['date'] = pd.to_datetime(df_transactions_store1['date'])
df_transactions_store1.set_index('date', inplace=True)
df_transactions_store1['sales'].plot()
plt.title('Time Plot')
plt.show()

df_transactions_store1
```

```python
df_transactions_store1['year'] = df_transactions_store1.index.year
df_transactions_store1['month'] = df_transactions_store1.index.month
df_transactions_store1['day'] = df_transactions_store1.index.day
df_transactions_store1['weekday'] = df_transactions_store1.index.weekday

df_transactions_store1
```

```python
df_transactions_store1.info()
```

Group per Yearly and Monthly

```python
# Membuat total transaksi per tahun dan bulan
df_transactions_store1.groupby(['year', 'month'])['sales'].sum()

# Membuat visualisasi data berdasarkan tahun dan bulan
plt.figure(figsize=(15, 5))
df_transactions_store1.groupby(['year', 'month'])['sales'].sum().plot()
plt.title('Total Transactions per Year and Month')
plt.show()
```

Autocorelation

```python
plt.figure(figsize=(15, 10))
autocorrelation_plot(df_transactions_store1['sales'])
plt.title('Auto Correlation Plot')
plt.show()
```

Decompose

```python
# Decompose Time Series



store20132015 = df_transactions_store1['2015' : '2016']
# store20132015 = df_transactions_store1[df_transactions_store1['year'] == 2013]


# Make fig size bigger
plt.rcParams['figure.figsize'] = (25, 15)
# Decompose time series
decomposition = seasonal_decompose(store20132015['sales'], model='additive', period=2)
fig = plt.figure()
fig = decomposition.plot()
```
